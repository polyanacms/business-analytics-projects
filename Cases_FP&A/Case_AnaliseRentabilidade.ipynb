# Análise de Rentabilidade por Produto e Cidade

## Objetivo
Analisar a rentabilidade da operação identificando quais produtos e regiões geram maior lucro, bem como aqueles que comprometem o resultado financeiro.

## Dor de Negócio
Empresas frequentemente enfrentam dificuldades em entender:
- Quais produtos são realmente lucrativos
- Se o crescimento de receita está gerando valor ou prejuízo
- Quais regiões operam com baixa rentabilidade

Isso pode levar a decisões equivocadas, como expandir operações pouco rentáveis ou manter produtos deficitários.

## Cenário
A empresa possui dados de:
- Receita por produto
- Custos diretos (COGS)
- Custos operacionais
- Localização (cidade)

O time de FP&A precisa:
- Avaliar lucro e margem
- Identificar produtos e regiões mais rentáveis
- Apoiar decisões estratégicas

## Abordagem
- Calcular lucro (Receita - Custos)
- Calcular margem (%)
- Analisar por produto e cidade
- Identificar top e bottom performers
- Visualizar distribuição de rentabilidade

## Insights Esperados
- Produtos mais e menos lucrativos
- Cidades com melhor performance
- Operações com margem negativa
- Oportunidades de melhoria

## Valor para o Negócio
- Otimização do portfólio de produtos
- Melhor alocação de recursos
- Suporte a decisões estratégicas (expandir, ajustar ou cortar)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Criando base simulada
n = 200

df = pd.DataFrame({
    'Produto': np.random.choice(['Produto A', 'Produto B', 'Produto C', 'Produto D'], n),
    'Cidade': np.random.choice(['SP', 'RJ', 'MG', 'PR'], n),
    'Receita': np.random.randint(5000, 20000, n),
    'Custo_Direto': np.random.randint(2000, 12000, n),
    'Custo_Operacional': np.random.randint(1000, 8000, n)
})

# 💰 Cálculo de lucro
df['Lucro'] = df['Receita'] - df['Custo_Direto'] - df['Custo_Operacional']

# 📊 Margem (%)
df['Margem'] = df['Lucro'] / df['Receita']

# 📊 Rentabilidade por produto
prod = df.groupby('Produto')[['Receita', 'Lucro']].sum()
prod['Margem'] = prod['Lucro'] / prod['Receita']
print(prod.sort_values(by='Lucro', ascending=False))

prod['Lucro'].plot(kind='bar', title='Lucro por Produto')
plt.show()

# 📊 Rentabilidade por cidade
cidade = df.groupby('Cidade')[['Receita', 'Lucro']].sum()
cidade['Margem'] = cidade['Lucro'] / cidade['Receita']
print(cidade.sort_values(by='Lucro', ascending=False))

cidade['Lucro'].plot(kind='bar', title='Lucro por Cidade')
plt.show()

# 📉 Identificar prejuízos
prejuizo = df[df['Lucro'] < 0]
print("Registros com prejuízo:")
print(prejuizo.head())

# 📊 Distribuição de margem
df['Margem'].hist(bins=20)
plt.title('Distribuição de Margem')
plt.show()
```
