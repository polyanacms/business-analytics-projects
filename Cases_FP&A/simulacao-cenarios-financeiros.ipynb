# CASE - Simulação de Cenários Financeiros

## Objetivo
Simular diferentes cenários financeiros para apoiar decisões estratégicas.

## Dor de Negócio
Sem simulação, decisões são tomadas:
- Sem prever impacto financeiro
- De forma reativa
- Com alto risco

## Cenário
A empresa quer entender:
- Impacto de aumento de custos
- Redução de despesas
- Crescimento de receita

## Abordagem
- Criar cenários:
  - Base
  - Otimista
  - Pessimista
- Simular variações (%)
- Comparar impacto no resultado

## Insights Esperados
- Sensibilidade do resultado a mudanças
- Riscos financeiros
- Oportunidades de ganho


## Valor para o Negócio
- Apoio à tomada de decisão
- Planejamento estratégico
- Redução de risco

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Base financeira inicial
base = {
    'Receita': 500000,
    'Custo_Variavel': 200000,
    'Custo_Fixo': 150000
}

df_base = pd.DataFrame([base])

# Cálculo do lucro base
df_base['Lucro'] = df_base['Receita'] - df_base['Custo_Variavel'] - df_base['Custo_Fixo']

print("Cenário Base:")
print(df_base)


# Definição dos cenários
cenarios = {
    'Base': {'Receita': 0.0, 'Custo_Variavel': 0.0, 'Custo_Fixo': 0.0},
    'Otimista': {'Receita': 0.10, 'Custo_Variavel': -0.05, 'Custo_Fixo': -0.02},
    'Pessimista': {'Receita': -0.10, 'Custo_Variavel': 0.08, 'Custo_Fixo': 0.05}
}

resultados = []


# Simulação dos cenários
for nome, variacao in cenarios.items():
    receita = base['Receita'] * (1 + variacao['Receita'])
    custo_var = base['Custo_Variavel'] * (1 + variacao['Custo_Variavel'])
    custo_fixo = base['Custo_Fixo'] * (1 + variacao['Custo_Fixo'])
    
    lucro = receita - custo_var - custo_fixo
    
    resultados.append({
        'Cenario': nome,
        'Receita': receita,
        'Custo_Variavel': custo_var,
        'Custo_Fixo': custo_fixo,
        'Lucro': lucro
    })

df_resultados = pd.DataFrame(resultados)


# Comparação dos cenários
print("\nComparação dos cenários:")
print(df_resultados)


# Gráfico de lucro por cenário
df_resultados.set_index('Cenario')['Lucro'].plot(kind='bar', title='Lucro por Cenário')
plt.axhline(0)
plt.show()


# Análise de sensibilidade simples
variacoes = np.arange(-0.2, 0.25, 0.05)
lucros = []

for v in variacoes:
    receita = base['Receita'] * (1 + v)
    lucro = receita - base['Custo_Variavel'] - base['Custo_Fixo']
    lucros.append(lucro)

df_sensibilidade = pd.DataFrame({
    'Variacao_Receita': variacoes,
    'Lucro': lucros
})

# Gráfico de sensibilidade
plt.plot(df_sensibilidade['Variacao_Receita'], df_sensibilidade['Lucro'])
plt.title('Sensibilidade do Lucro à Receita')
plt.xlabel('Variação da Receita')
plt.ylabel('Lucro')
plt.show()
```
