# Análise de Eficiência Operacional (Custo por Unidade)

## Objetivo
Avaliar a eficiência operacional através do custo por unidade (ex: custo por loja, cliente ou operação).

## Dor de Negócio
Empresas não sabem:
- Quais unidades são mais eficientes
- Onde estão os maiores custos relativos
- Como comparar performance entre regiões

## Cenário
A empresa possui:
- Despesas por cidade/unidade
- Volume de operação (ex: número de clientes, vendas ou atendimentos)

O objetivo é:
- Calcular custo por unidade
- Comparar performance entre regiões

## Abordagem
- Calcular custo total por unidade
- Dividir pelo volume
- Rankear eficiência
- Identificar outliers

## Insights Esperados
- Unidades mais eficientes
- Unidades com alto custo relativo
- Oportunidades de melhoria

## Valor para o Negócio
- Otimização operacional
- Melhor alocação de recursos
- Benchmark interno

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Criando base simulada
n = 150

df = pd.DataFrame({
    'Cidade': np.random.choice(['SP', 'RJ', 'MG', 'PR', 'RS'], n),
    'Unidade': np.random.choice(['Loja A', 'Loja B', 'Loja C', 'Loja D'], n),
    'Custo_Total': np.random.randint(50000, 200000, n),
    'Volume': np.random.randint(300, 2000, n)  # clientes, vendas ou atendimentos
})

# Cálculo de custo por unidade
df['Custo_por_Unidade'] = df['Custo_Total'] / df['Volume']

# Visualização inicial
print("Amostra dos dados:")
print(df.head())

# Agregação por cidade
cidade = df.groupby('Cidade')[['Custo_Total', 'Volume']].sum()
cidade['Custo_por_Unidade'] = cidade['Custo_Total'] / cidade['Volume']

print("\nEficiência por cidade:")
print(cidade.sort_values(by='Custo_por_Unidade'))

# Ranking de eficiência (menor custo = mais eficiente)
cidade_sorted = cidade.sort_values(by='Custo_por_Unidade')

# Gráfico: custo por unidade por cidade
cidade_sorted['Custo_por_Unidade'].plot(kind='bar', title='Custo por Unidade por Cidade')
plt.show()

# Agregação por unidade (loja)
unidade = df.groupby('Unidade')[['Custo_Total', 'Volume']].sum()
unidade['Custo_por_Unidade'] = unidade['Custo_Total'] / unidade['Volume']

print("\nEficiência por unidade:")
print(unidade.sort_values(by='Custo_por_Unidade'))

# Identificação de outliers (top e bottom)
top_ineficientes = df.sort_values(by='Custo_por_Unidade', ascending=False).head(10)
top_eficientes = df.sort_values(by='Custo_por_Unidade').head(10)

print("\nTop 10 unidades menos eficientes:")
print(top_ineficientes[['Cidade', 'Unidade', 'Custo_Total', 'Volume', 'Custo_por_Unidade']])

print("\nTop 10 unidades mais eficientes:")
print(top_eficientes[['Cidade', 'Unidade', 'Custo_Total', 'Volume', 'Custo_por_Unidade']])

# Distribuição do custo por unidade
df['Custo_por_Unidade'].hist(bins=20)
plt.title('Distribuição do Custo por Unidade')
plt.show()

# Análise cruzada (Cidade x Unidade)
pivot = df.pivot_table(values='Custo_por_Unidade', index='Cidade', columns='Unidade', aggfunc='mean')

print("\nMatriz de eficiência (Cidade x Unidade):")
print(pivot)
```
