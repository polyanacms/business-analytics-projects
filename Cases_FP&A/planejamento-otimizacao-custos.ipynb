# Planejamento e Otimização de Custos Operacionais por Capacidade

## Objetivo
Planejar e analisar os custos operacionais com base na capacidade produtiva, garantindo eficiência na alocação de recursos e aderência ao orçamento.

## Dor de Negócio
Empresas que trabalham com capacidade (ex: atendimento, produção, operações) enfrentam desafios como:
- Superdimensionamento de recursos (custo alto e ociosidade)
- Subdimensionamento (queda de qualidade e sobrecarga)
- Falta de previsibilidade de custos operacionais
- Dificuldade em ajustar planejamento ao longo do tempo

## Cenário
A empresa possui:
- Demanda projetada (ex: número de atendimentos, turmas, operações)
- Capacidade produtiva por recurso (ex: pessoas, equipes)
- Custo por recurso

O desafio é:
- Planejar quantos recursos são necessários
- Calcular o custo total
- Monitorar eficiência (capacidade vs demanda)
- Identificar desperdícios ou gargalos

## Abordagem
- Estimar capacidade necessária com base na demanda
- Calcular custo total planejado
- Comparar capacidade planejada vs utilizada
- Identificar ociosidade ou sobrecarga
- Simular ajustes para otimização

## Insights Esperados
- Nível de eficiência operacional
- Identificação de excesso ou falta de recursos
- Impacto financeiro da má alocação
- Oportunidades de otimização de custo

## Valor para o Negócio
- Redução de custos operacionais
- Melhor planejamento de recursos
- Aumento de eficiência
- Base para decisões estratégicas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Base simulada
n = 100

df = pd.DataFrame({
    'Periodo': pd.date_range(start='2024-01-01', periods=n, freq='W'),
    'Regiao': np.random.choice(['Sudeste', 'Sul', 'Nordeste'], n),
    'Demanda': np.random.randint(200, 1000, n),  # volume esperado
    'Capacidade_por_Recurso': np.random.randint(80, 120, n),  # quanto cada recurso atende
    'Custo_por_Recurso': np.random.randint(3000, 7000, n)
})


# Planejamento de recursos
df['Recursos_Necessarios'] = np.ceil(df['Demanda'] / df['Capacidade_por_Recurso'])


# Custo total planejado
df['Custo_Total'] = df['Recursos_Necessarios'] * df['Custo_por_Recurso']


# Eficiência operacional
df['Capacidade_Total'] = df['Recursos_Necessarios'] * df['Capacidade_por_Recurso']
df['Ociosidade'] = df['Capacidade_Total'] - df['Demanda']


# Visão geral
print("Amostra dos dados:")
print(df.head())


# Análise por região
regiao = df.groupby('Regiao')[['Demanda', 'Capacidade_Total', 'Custo_Total']].sum()
regiao['Eficiencia_%'] = regiao['Demanda'] / regiao['Capacidade_Total']

print("\nResumo por região:")
print(regiao)


# Gráfico de eficiência
regiao['Eficiencia_%'].plot(kind='bar', title='Eficiência por Região')
plt.show()


# Identificação de desperdício
top_ociosidade = df.sort_values(by='Ociosidade', ascending=False).head(10)

print("\nTop períodos com maior ociosidade:")
print(top_ociosidade[['Periodo', 'Regiao', 'Ociosidade', 'Custo_Total']])


# Evolução de custo
custo_periodo = df.groupby('Periodo')['Custo_Total'].sum()

custo_periodo.plot(title='Evolução do Custo ao Longo do Tempo')
plt.show()


# Simulação de otimização
# Redução de 10% na ociosidade
df['Custo_Otimizado'] = df['Custo_Total'] * 0.9

economia = df['Custo_Total'].sum() - df['Custo_Otimizado'].sum()

print("\nEconomia potencial com otimização:", economia)
```
