# Forecast e Análise de Tendência de Despesas

## Objetivo
Construir uma análise de tendência e previsão de despesas para apoiar o planejamento financeiro e melhorar a previsibilidade do negócio.

## Dor de Negócio
Sem previsões estruturadas, empresas enfrentam:
- Falta de previsibilidade no fechamento financeiro
- Dificuldade no planejamento orçamentário
- Decisões baseadas apenas em histórico, sem projeção

## Cenário
O time de FP&A possui dados históricos mensais de despesas, mas não possui um modelo estruturado de forecast.

O objetivo é:
- Entender o comportamento histórico
- Identificar tendências
- Criar uma previsão inicial

## Abordagem
- Construção de série temporal
- Suavização com média móvel
- Comparação entre valores reais e tendência
- Projeção futura simples

## Insights Esperados
- Tendência de crescimento ou redução
- Sazonalidade (se houver)
- Base para tomada de decisão futura

## Valor para o Negócio
- Melhor previsibilidade financeira
- Apoio ao planejamento estratégico
- Redução de surpresas no resultado

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Criando série temporal mais longa
meses = pd.date_range(start='2022-01-01', periods=36, freq='M')
valores = np.random.normal(loc=12000, scale=2000, size=36)

df = pd.DataFrame({
    'Mes': meses,
    'Despesa': valores
})

# 📊 Média móvel
df['MM_3'] = df['Despesa'].rolling(3).mean()
df['MM_6'] = df['Despesa'].rolling(6).mean()

# 📈 Tendência simples (regressão)
x = np.arange(len(df))
coef = np.polyfit(x, df['Despesa'], 1)
trend = coef[0] * x + coef[1]

# Plot
plt.figure()
plt.plot(df['Mes'], df['Despesa'], label='Real')
plt.plot(df['Mes'], df['MM_3'], label='MM 3 meses')
plt.plot(df['Mes'], df['MM_6'], label='MM 6 meses')
plt.plot(df['Mes'], trend, label='Tendência', linestyle='--')

plt.legend()
plt.title('Forecast e Tendência de Despesas')
plt.xticks(rotation=45)
plt.show()

# 📊 Projeção simples (próximos 3 meses)
future_x = np.arange(len(df), len(df)+3)
forecast = coef[0] * future_x + coef[1]

print("Forecast próximos 3 meses:", forecast)
```
