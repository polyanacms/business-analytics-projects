# Análise Completa de Despesas (FP&A + Data Analytics)

## Objetivo
Realizar uma análise aprofundada das despesas para identificar padrões, concentração de custos, variações ao longo do tempo e oportunidades de otimização financeira.

## Dor de Negócio
Empresas com múltiplas unidades e categorias de despesas enfrentam desafios como:
- Falta de visibilidade sobre os principais drivers de custo
- Dificuldade em identificar desvios relevantes
- Baixa capacidade de priorizar ações de eficiência

## Cenário
A empresa possui operações em diversas cidades e registra despesas nas seguintes categorias:
- Aluguel
- Energia
- TI
- Manutenção
- Marketing

O time de FP&A precisa entender:
- Quais categorias representam maior impacto financeiro
- Como as despesas evoluem ao longo do tempo
- Diferenças entre cidades
- Possíveis anomalias ou picos de gastos

## Abordagem
- Criação de base simulada com múltiplas dimensões
- Análise por categoria, cidade e período
- Identificação de concentração de custos (Pareto)
- Análise de tendência temporal
- Identificação de picos e variações relevantes

## Insights Esperados
- Top categorias responsáveis pela maior parte dos custos
- Cidades com maior consumo de recursos
- Tendências de crescimento ou estabilidade
- Identificação de possíveis ineficiências

## Valor para o Negócio
- Apoio à tomada de decisão estratégica
- Identificação de alavancas de redução de custo
- Melhoria no controle orçamentário
- Base para construção de forecast

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Criando base mais realista
n = 300

df = pd.DataFrame({
    'Data': pd.date_range(start='2023-01-01', periods=n, freq='D'),
    'Categoria': np.random.choice(['Aluguel', 'Energia', 'TI', 'Manutenção', 'Marketing'], n),
    'Cidade': np.random.choice(['SP', 'RJ', 'MG', 'PR'], n),
    'Valor': np.random.randint(500, 10000, n)
})

df['Mes'] = df['Data'].dt.to_period('M')

# 📊 1. Despesa total por categoria
cat = df.groupby('Categoria')['Valor'].sum().sort_values(ascending=False)
cat.plot(kind='bar', title='Despesas por Categoria')
plt.show()

# 📊 2. Pareto (80/20)
cat_perc = cat / cat.sum()
cat_perc.cumsum().plot(title='Curva de Pareto')
plt.show()

# 📊 3. Despesa por cidade
cidade = df.groupby('Cidade')['Valor'].sum()
cidade.plot(kind='bar', title='Despesas por Cidade')
plt.show()

# 📊 4. Evolução mensal
mensal = df.groupby('Mes')['Valor'].sum()
mensal.plot(title='Evolução Mensal')
plt.show()

# 📊 5. Identificação de picos
print(df.sort_values(by='Valor', ascending=False).head(10))
```
