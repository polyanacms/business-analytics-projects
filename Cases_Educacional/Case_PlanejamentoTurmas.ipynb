# Planejamento de Capacidade de Turmas

## Objetivo
Planejar a quantidade de turmas com base na demanda de alunos.

## Dor de Negócio
- Turmas lotadas ou ociosas
- Ineficiência operacional
- Custos desnecessários

## Cenário
Dados:
- Número de alunos
- Capacidade por turma

## Abordagem
- Calcular número de turmas necessárias
- Avaliar ociosidade

## Insights Esperados
- Necessidade de turmas
- Ociosidade
- Eficiência

## Valor para o Negócio
- Redução de custos
- Melhor uso de recursos

```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 100

df = pd.DataFrame({
    'Curso': np.random.choice(['A','B','C'], n),
    'Alunos': np.random.randint(20, 120, n),
    'Capacidade_Turma': 40
})

df['Turmas_Necessarias'] = np.ceil(df['Alunos'] / df['Capacidade_Turma'])
df['Capacidade_Total'] = df['Turmas_Necessarias'] * df['Capacidade_Turma']
df['Ociosidade'] = df['Capacidade_Total'] - df['Alunos']

print(df.head())
```
