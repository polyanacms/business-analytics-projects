# Análise de Viabilidade Financeira de Turmas Acadêmicas

## Objetivo
Avaliar a viabilidade financeira da abertura de turmas acadêmicas, considerando receita, custos docentes e despesas operacionais.

## Dor de Negócio
Instituições de ensino enfrentam desafios como:
- Abrir turmas com baixa rentabilidade
- Falta de integração entre áreas acadêmicas e financeiras
- Dificuldade em prever o resultado antes da execução
Isso pode gerar prejuízo operacional e má alocação de recursos.

## Cenário
A instituição possui dados de:
- Número de alunos matriculados por turma
- Mensalidade
- Carga horária do curso
- Custo por hora docente
- Custos operacionais

O time precisa decidir:
- Quais turmas devem ser abertas
- Quais precisam de ajustes (preço ou capacidade)
- Quais devem ser canceladas

## Abordagem
- Calcular receita total por turma
- Calcular custo docente com base na carga horária
- Somar custos operacionais
- Calcular lucro e margem
- Classificar viabilidade

## Insights Esperados
- Turmas lucrativas vs deficitárias
- Ponto de equilíbrio (break-even)
- Sensibilidade ao número de alunos
- Apoio à decisão de abertura ou cancelamento

## Valor para o Negócio
- Melhor alocação de recursos
- Redução de prejuízos
- Aumento da eficiência operacional
- Integração entre áreas acadêmica e financeira

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Base de turmas
n = 10

df = pd.DataFrame({
    'Turma': [f'T{i}' for i in range(1, n+1)],
    'Alunos': np.random.randint(10, 60, n),
    'Mensalidade': np.random.randint(500, 1500, n),
    'Carga_Horaria': np.random.randint(40, 120, n),
    'Custo_Hora_Docente': np.random.randint(50, 120, n),
    'Custo_Operacional': np.random.randint(5000, 15000, n)
})

# Receita (considerando 6 meses de curso)
df['Receita'] = df['Alunos'] * df['Mensalidade'] * 6

# Custo docente
df['Custo_Docente'] = df['Carga_Horaria'] * df['Custo_Hora_Docente']

# Custo total
df['Custo_Total'] = df['Custo_Docente'] + df['Custo_Operacional']

# Resultado
df['Lucro'] = df['Receita'] - df['Custo_Total']
df['Margem'] = df['Lucro'] / df['Receita']

# Classificação
def classificar(lucro):
    if lucro < 0:
        return 'Inviavel'
    elif lucro < 10000:
        return 'Atenção'
    else:
        return 'Viavel'

df['Status'] = df['Lucro'].apply(classificar)

print(df)

# Resumo
print("\nResumo por status:")
print(df['Status'].value_counts())

# Gráfico
df.set_index('Turma')['Lucro'].plot(kind='bar')
plt.title('Lucro por Turma')
plt.show()
```
