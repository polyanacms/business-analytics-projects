# Análise de Evasão de Alunos

## Objetivo
Identificar fatores que influenciam a evasão de alunos, permitindo ações preventivas.

## Dor de Negócio
Instituições de ensino enfrentam:
- Alta evasão de alunos
- Perda de receita
- Dificuldade em identificar causas

## Cenário
A instituição possui dados de:
- Frequência
- Notas
- Mensalidade
- Status (ativo ou evadido)

## Abordagem
- Analisar relação entre frequência e evasão
- Avaliar desempenho acadêmico
- Identificar padrões de risco

## Insights Esperados
- Perfil de alunos com maior risco
- Relação entre nota e evasão
- Indicadores críticos

## Valor para o Negócio
- Redução da evasão
- Aumento de retenção
- Melhor planejamento acadêmico

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n = 300

df = pd.DataFrame({
    'Frequencia': np.random.randint(50, 100, n),
    'Nota': np.random.uniform(4, 10, n),
    'Mensalidade': np.random.randint(500, 2000, n),
    'Evadido': np.random.choice([0,1], n, p=[0.75,0.25])
})

print(df.head())

evasao = df.groupby(pd.cut(df['Frequencia'], bins=5))['Evadido'].mean()
evasao.plot(kind='bar', title='Evasão por Frequência')
plt.show()
```
