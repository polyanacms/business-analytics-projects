# Análise de Rentabilidade por Curso

## Objetivo
Avaliar quais cursos geram maior retorno financeiro.

## Dor de Negócio
Instituições enfrentam:
- Cursos pouco rentáveis
- Custos elevados
- Falta de visibilidade financeira

## Cenário
Dados disponíveis:
- Receita por curso
- Custo operacional
- Número de alunos

## Abordagem
- Calcular lucro
- Calcular margem
- Comparar cursos

## Insights Esperados
- Cursos mais lucrativos
- Cursos deficitários
- Oportunidades de otimização

## Valor para o Negócio
- Melhor alocação de recursos
- Otimização do portfólio

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

cursos = ['Administração', 'Engenharia', 'Direito', 'TI']

df = pd.DataFrame({
    'Curso': cursos,
    'Receita': np.random.randint(100000, 500000, len(cursos)),
    'Custo': np.random.randint(50000, 300000, len(cursos))
})

df['Lucro'] = df['Receita'] - df['Custo']
df['Margem'] = df['Lucro'] / df['Receita']

print(df)

df.set_index('Curso')['Lucro'].plot(kind='bar', title='Lucro por Curso')
plt.show()
```
