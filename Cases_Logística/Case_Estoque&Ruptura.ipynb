```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n = 100

df = pd.DataFrame({
    'Produto': np.random.choice(['A', 'B', 'C', 'D'], n),
    'Estoque_Atual': np.random.randint(0, 500, n),
    'Consumo_Medio': np.random.randint(10, 100, n)
})

# Dias de cobertura
df['Dias_Cobertura'] = df['Estoque_Atual'] / df['Consumo_Medio']

# Classificação
def classificar(dias):
    if dias < 5:
        return 'Ruptura'
    elif dias > 30:
        return 'Excesso'
    else:
        return 'Saudável'

df['Status'] = df['Dias_Cobertura'].apply(classificar)

print(df.head())

# Distribuição
status = df['Status'].value_counts()
print("\nStatus do estoque:")
print(status)

# Gráfico
status.plot(kind='bar', title='Classificação de Estoque')
plt.show()
```

# Análise de Estoque e Risco de Ruptura

## Objetivo
Avaliar o nível de estoque para identificar riscos de ruptura ou excesso, garantindo equilíbrio operacional.

## Dor de Negócio
Empresas enfrentam:
- Falta de produtos (ruptura)
- Excesso de estoque (capital parado)
- Falta de visibilidade sobre cobertura de estoque
Isso impacta vendas, custos e eficiência financeira.


## Cenário
A empresa possui dados de:
- Estoque atual
- Consumo médio de produtos

O time precisa:
- Calcular cobertura de estoque
- Identificar riscos
- Apoiar decisões de reposição

## Abordagem
- Calcular dias de cobertura (estoque / consumo)
- Classificar produtos:
  - Ruptura
  - Saudável
  - Excesso
- Analisar distribuição por produto
- Identificar outliers


## Insights Esperados
- Produtos com risco de ruptura
- Produtos com excesso de estoque
- Distribuição da saúde do estoque
- Oportunidades de ajuste


## Valor para o Negócio
- Redução de perdas por ruptura
- Otimização do capital de giro
- Melhor planejamento de compras
- Aumento da eficiência logística
