```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Base simulada
n = 200

df = pd.DataFrame({
    'Origem': np.random.choice(['SP', 'RJ', 'MG'], n),
    'Destino': np.random.choice(['PR', 'SC', 'RS'], n),
    'Distancia_km': np.random.randint(100, 1500, n),
    'Peso_kg': np.random.randint(100, 1000, n),
    'Custo_Frete': np.random.randint(500, 5000, n)
})

# Custo por km e por kg
df['Custo_por_km'] = df['Custo_Frete'] / df['Distancia_km']
df['Custo_por_kg'] = df['Custo_Frete'] / df['Peso_kg']

print(df.head())

# Análise por rota
rota = df.groupby(['Origem', 'Destino'])['Custo_Frete'].mean().sort_values(ascending=False)
print("\nCusto médio por rota:")
print(rota)

# Gráfico
rota.head(10).plot(kind='bar', title='Top rotas mais caras')
plt.show()
```

# Análise de Custos de Transporte por Rota

## Objetivo
Analisar os custos logísticos de transporte, identificando rotas mais onerosas e oportunidades de otimização.

## Dor de Negócio
Empresas frequentemente enfrentam dificuldades em:
- Identificar rotas com custos elevados
- Entender eficiência do frete em relação à distância e peso
- Controlar gastos logísticos crescentes
Isso pode levar a margens reduzidas e decisões ineficientes na distribuição.

## Cenário
A empresa possui dados de:
- Origem e destino das entregas
- Distância percorrida
- Peso transportado
- Custo do frete

O time precisa:
- Avaliar eficiência das rotas
- Identificar gargalos de custo
- Apoiar decisões de otimização logística

## Abordagem
- Calcular custo por km
- Calcular custo por kg
- Agrupar por rota (origem x destino)
- Identificar rotas mais caras
- Comparar eficiência entre rotas

## Insights Esperados
- Rotas com maior custo logístico
- Ineficiências no transporte
- Oportunidades de renegociação ou otimização
- Melhor alocação de cargas

## Valor para o Negócio
- Redução de custos logísticos
- Aumento de eficiência operacional
- Melhor planejamento de transporte
- Suporte à estratégia de distribuição
