```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n = 150

df = pd.DataFrame({
    'Data_Pedido': pd.date_range(start='2024-01-01', periods=n, freq='D'),
    'Prazo_Prometido': np.random.randint(2, 7, n),
    'Tempo_Real': np.random.randint(1, 10, n),
    'Regiao': np.random.choice(['Sudeste', 'Sul', 'Nordeste'], n)
})

# Lead time real
df['Atraso'] = df['Tempo_Real'] - df['Prazo_Prometido']

print(df.head())

# Análise por região
regiao = df.groupby('Regiao')['Atraso'].mean()
print("\nAtraso médio por região:")
print(regiao)

# Percentual de entregas atrasadas
df['Atrasado'] = df['Atraso'] > 0
taxa_atraso = df['Atrasado'].mean()

print("\nTaxa de atraso:", taxa_atraso)

# Gráfico
regiao.plot(kind='bar', title='Atraso médio por região')
plt.show()
```

# Análise de Lead Time e Atrasos nas Entregas

## Objetivo
Avaliar o tempo de entrega e identificar atrasos, melhorando o nível de serviço ao cliente.

## Dor de Negócio
Empresas enfrentam desafios como:
- Entregas fora do prazo
- Falta de previsibilidade logística
- Insatisfação de clientes
Isso impacta diretamente a experiência do cliente e a reputação da empresa.


## Cenário
A empresa possui dados de:
- Data do pedido
- Prazo prometido
- Tempo real de entrega
- Região de entrega

O objetivo é:
- Medir o nível de atraso
- Identificar regiões críticas
- Melhorar o desempenho logístico

## Abordagem
- Calcular atraso (tempo real - prazo prometido)
- Identificar entregas atrasadas
- Calcular taxa de atraso
- Analisar por região
- Avaliar distribuição dos tempos de entrega

## Insights Esperados
- Regiões com maior atraso
- Percentual de entregas fora do prazo
- Gargalos operacionais
- Oportunidades de melhoria logística

## Valor para o Negócio
- Melhoria do nível de serviço (SLA)
- Aumento da satisfação do cliente
- Redução de penalidades/logística reversa
- Melhor gestão de prazos
