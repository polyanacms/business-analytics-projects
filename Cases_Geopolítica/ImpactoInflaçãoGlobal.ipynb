# Análise de Impacto da Inflação Global

## Objetivo
Avaliar impacto da inflação em diferentes países.

## Dor de Negócio
Empresas globais enfrentam:
- Aumento de custos
- Perda de poder de compra
- Risco financeiro

## Cenário
Dados:
- País
- Taxa de inflação
- Custo base

## Abordagem
- Ajustar custos pela inflação
- Comparar países

## Insights Esperados
- Países mais impactados
- Risco inflacionário

## Valor para o Negócio
- Melhor planejamento internacional

```python
import pandas as pd
import numpy as np

np.random.seed(42)

df = pd.DataFrame({
    'Pais': ['Brasil','EUA','Argentina','Alemanha'],
    'Inflacao': [0.06,0.03,0.12,0.02],
    'Custo_Base': [100000,100000,100000,100000]
})

df['Custo_Ajustado'] = df['Custo_Base'] * (1 + df['Inflacao'])

print(df)
```
