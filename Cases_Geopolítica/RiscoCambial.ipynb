# Análise de Risco Cambial

## Objetivo
Avaliar impacto da variação cambial.

## Dor de Negócio
- Volatilidade cambial
- Impacto em receitas e custos

## Cenário
Dados:
- Moeda
- Taxa de câmbio

## Abordagem
- Simular variação cambial
- Avaliar impacto financeiro

## Insights Esperados
- Sensibilidade ao câmbio

## Valor para o Negócio
- Gestão de risco

```python
import pandas as pd
import numpy as np

np.random.seed(42)

df = pd.DataFrame({
    'Moeda': ['USD','EUR','CNY'],
    'Taxa_Atual': [5.0,5.5,0.7]
})

variacao = 0.1

df['Taxa_Simulada'] = df['Taxa_Atual'] * (1 + variacao)

print(df)
```
