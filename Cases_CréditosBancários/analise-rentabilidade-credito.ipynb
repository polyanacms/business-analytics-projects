# Análise de Rentabilidade da Carteira de Crédito

## Objetivo
Avaliar a rentabilidade da carteira considerando receitas de juros e perdas com inadimplência.

## Dor de Negócio
Instituições financeiras enfrentam:
- Dificuldade em equilibrar risco e retorno
- Perdas com inadimplência
- Falta de visibilidade sobre rentabilidade real

## Cenário
A empresa possui dados de:
- Valor emprestado
- Taxa de juros
- Status de inadimplência

O objetivo é:
- Calcular retorno da carteira
- Avaliar impacto das perdas
- Medir rentabilidade líquida

## Abordagem
- Calcular receita de juros
- Estimar perdas por inadimplência
- Calcular lucro líquido
- Analisar distribuição de resultados

## Insights Esperados
- Rentabilidade da carteira
- Impacto da inadimplência no resultado
- Perfis mais lucrativos
- Oportunidades de otimização

## Valor para o Negócio
- Melhor gestão da carteira
- Equilíbrio entre risco e retorno
- Aumento da rentabilidade

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n = 200

df = pd.DataFrame({
    'Emprestimo': np.random.randint(1000, 20000, n),
    'Taxa_Juros': np.random.uniform(0.02, 0.08, n),
    'Inadimplente': np.random.choice([0, 1], n, p=[0.85, 0.15])
})

# Receita de juros
df['Receita'] = df['Emprestimo'] * df['Taxa_Juros']

# Perdas
df['Perda'] = df['Emprestimo'] * df['Inadimplente']

# Lucro
df['Lucro'] = df['Receita'] - df['Perda']

print(df.head())

# Resultado total
lucro_total = df['Lucro'].sum()
print("\nLucro total da carteira:", lucro_total)

# Distribuição
df['Lucro'].hist(bins=20)
plt.title('Distribuição de Lucro')
plt.show()
```
