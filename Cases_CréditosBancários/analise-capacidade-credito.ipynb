# Análise de Capacidade de Crédito

## Objetivo
Avaliar se o limite de crédito concedido está adequado à capacidade financeira do cliente.

## Dor de Negócio
Empresas enfrentam:
- Concessão de crédito acima da capacidade do cliente
- Aumento do risco de inadimplência
- Falta de critérios claros para definição de limite

## Cenário
A empresa possui dados de:
- Renda mensal
- Limite de crédito concedido

O objetivo é:
- Avaliar comprometimento da renda
- Identificar clientes em risco
- Apoiar definição de limites mais adequados

## Abordagem
- Calcular percentual de comprometimento da renda
- Classificar níveis de risco
- Analisar distribuição dos perfis
- Identificar clientes com alto risco

## Insights Esperados
- Percentual de clientes com alto comprometimento
- Distribuição dos níveis de risco
- Identificação de clientes superexpostos
- Oportunidades de ajuste de limite

## Valor para o Negócio
- Redução da inadimplência
- Melhor gestão de risco
- Política de crédito mais eficiente

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n = 250

df = pd.DataFrame({
    'Renda': np.random.randint(2000, 15000, n),
    'Limite_Credito': np.random.randint(1000, 20000, n)
})

# Comprometimento da renda
df['Comprometimento'] = df['Limite_Credito'] / df['Renda']

# Classificação de risco
def classificar(x):
    if x < 0.3:
        return 'Saudavel'
    elif x < 0.6:
        return 'Moderado'
    else:
        return 'Alto Risco'

df['Classificacao'] = df['Comprometimento'].apply(classificar)

print(df.head())

# Distribuição
dist = df['Classificacao'].value_counts()
print("\nClassificação de risco:")
print(dist)

# Gráfico
dist.plot(kind='bar', title='Perfil de Risco de Crédito')
plt.show()
```
