# Análise de Risco de Inadimplência

## Objetivo
Identificar clientes com maior risco de inadimplência, apoiando decisões de concessão de crédito.

## Dor de Negócio
Instituições financeiras enfrentam desafios como:
- Alta taxa de inadimplência
- Dificuldade em prever risco de crédito
- Perdas financeiras com clientes de alto risco
Isso impacta diretamente o resultado financeiro e a sustentabilidade do negócio.


## Cenário
A empresa possui dados de:
- Score de crédito
- Renda do cliente
- Valor do empréstimo
- Histórico de pagamento

O objetivo é:
- Identificar padrões de inadimplência
- Classificar clientes por risco
- Apoiar decisões de crédito


## Abordagem
- Analisar relação entre score e inadimplência
- Agrupar clientes por faixa de score
- Calcular taxa de inadimplência por grupo
- Identificar perfis de maior risco

## Insights Esperados
- Relação entre score e inadimplência
- Faixas de score com maior risco
- Perfis críticos de clientes
- Oportunidades de ajuste na concessão

## Valor para o Negócio
- Redução de perdas financeiras
- Melhoria na política de crédito
- Maior segurança na concessão

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n = 300

df = pd.DataFrame({
    'Score': np.random.randint(300, 900, n),
    'Renda': np.random.randint(2000, 15000, n),
    'Emprestimo': np.random.randint(1000, 20000, n),
    'Inadimplente': np.random.choice([0, 1], n, p=[0.8, 0.2])
})

print(df.head())

# Taxa geral de inadimplência
taxa = df['Inadimplente'].mean()
print("\nTaxa de inadimplência:", taxa)

# Inadimplência por faixa de score
df['Faixa_Score'] = pd.cut(df['Score'], bins=5)
score = df.groupby('Faixa_Score')['Inadimplente'].mean()

print("\nInadimplência por faixa de score:")
print(score)

# Gráfico
score.plot(kind='bar', title='Inadimplência por Faixa de Score')
plt.show()
```
